### Create functions to investigate columns

```python
def find_mean_columns(category,b):
    list_mean = []
    for i in category:
        list_mean.append(df[i].mean()) 
        
    dict_category = {} 
    for key in category: 
        for value in list_mean: 
            dict_category[key] = value 
            list_mean.remove(value) 
            break
            
    print([i for i, occurrences in dict_category.items() if occurrences >= b])
# use for numeric flag columns to determine which values exist more commonly in the data set
# drop columns that do not occur as frequently in the data set
```

```python
def find_difference(category):
    df_category=df[category]
    df_category_mean = df_category.groupby('disposition').mean()
    df_category_mean = df_category_mean.transpose()
    df_category_mean['diff'] = df_category_mean[1] - df_category_mean[0]
    print(df_category_mean.nlargest(10,['diff']))
    print(df_category_mean.nsmallest(10,['diff']))
# use for numeric flag columns to determine which values exist more commonly by disposition
```

```python
def find_max_count(category):
    list_max = []
    for i in category:
        list_max.append(df[i].max())
    print(list_max)
    
    dict_category = {} 
    for key in category: 
        for value in list_max: 
            dict_category[key] = value 
            list_max.remove(value) 
            break
    print(dict_category)
```
